```python
import pandas as pd

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')
dados
```

```python
dados.columns = [coluna.replace('-','') for coluna in dados.columns]
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True)
dados
```

# Desafio 1:
*Ordenar o gr√°fico countplot*

```python
cod_compostos = dados['composto'].value_counts().index[0:5]
cod_compostos
```

```python
dados.query('composto in @cod_compostos')
```

`order = cod_compostos`

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(8, 6))
ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'), order = cod_compostos, edgecolor=(0,0,0))
ax.set_title('Top 5 Compostos')
plt.show()
```

#Desafio 2:
*Melhorar a visualiza√ß√£o alterando tamanho da fonte*

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(8, 6))
ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'), order = cod_compostos, edgecolor=(0,0,0))
ax.set(title='Top 5 Compostos', xlabel = 'Compostos', ylabel = 'Quantidade de Vezes Utilizada')
plt.show()
```
